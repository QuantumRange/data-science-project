```python
import polars as pl
import matplotlib.pyplot as plt
```

```python
(
    pl
    .scan_parquet("/mnt/Fast2T/data/new/stage_5/*.parquet")
    .explode("stems")
    .unnest("stems")
    .group_by("is_ai", "stem")
    .agg(pl.col("occurrence").sum())
    .with_columns(p=pl.col("occurrence") / pl.col("occurrence").sum().over("is_ai"))
    .sink_parquet(
        "/mnt/Fast2T/data/word_distribution.parquet",
        compression="zstd",
        compression_level=21
    )
)
```
