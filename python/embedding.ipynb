```
import glob
from pathlib import Path

import polars as pl
```

### Stem Embedding
Stem embedding works by taking the top $K$ words and encoding the frequency of them appearing.

```
WEBSITE_FILES = sorted(glob.glob("/mnt/Fast2T/data/link/*.parquet"))
TARGET_RAW = Path("/mnt/Fast2T/data/ai-training-set/raw/")
```

```
for f in WEBSITE_FILES:
    df = (pl.read_parquet(f)
          .filter(pl.col("lang") == "ENGLISH")
          .select(pl.col("id"), pl.col("text").str.split(' ').alias("term"))
          .explode("term")
          .filter(pl.col("term").str.contains(r'\d|\.|"').not_())
          .group_by("id", "term")
          .len()
          .sort("len", descending=True))
    break

df
```
