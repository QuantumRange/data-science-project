```python
import json
import polars as pl
from pathlib import Path
from tqdm import tqdm

posts_dir = Path("/home/qr/Downloads/moltbook_data/data/posts")

data = []

for post_file in tqdm(posts_dir.glob("*.json")):
    post = json.loads(post_file.read_text())

    if "content" in post["post"]:
        data.append(post["post"]["content"])

    if "comments" in post:
        for comment in post["comments"]:
            data.append(comment["content"])

pl.DataFrame({
    "text": data,
}).write_parquet("/mnt/Fast2T/data/moltbook.parquet", compression="zstd", compression_level=22)
```
